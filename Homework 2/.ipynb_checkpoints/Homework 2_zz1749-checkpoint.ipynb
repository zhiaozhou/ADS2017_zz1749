###  Regress over these proportions (the ones mentioned in part b, leave out "No Insurance Coverage") to predict income -- What interesting results do we have here? Briefly state what you've obtained

### Visualize the observed income versus the predicted income, like what we did in the lab session. Does this fitted line look great or not? If you're not sure, check the correlation matrix and make a guess.

```python
ex2['TotalIncome'] = ex2.IncomePerCapita * ex2.TotalPop
```

```python
train, valid = train_test_split(ex2, test_size = 0.3)
```

```python
lm1_ex2 = smf.ols(formula = 'TotalIncome~TotalPop+PrivateOnly+PublicOnly+WithBothPublicAndPrivate',data=train).fit()
```

```python
def modelEval(lm, key = 'TotalIncome'):
    lmy = lm.predict(valid)
    y_err = lmy - valid[key]
    y_norm = valid[key]-mean(valid[key])
    R2 = 1-y_err.dot(y_err)/y_norm.dot(y_norm)  # R2 w.r.t. validation set
    print('Validation R^2 is %f'%R2)
    print(lm.summary())
    plt.plot(lmy, lmy, 'b-')
    plt.plot(lmy, valid[key],'or')
```

```python
modelEval(lm1_ex2)
```

### The p-value of PublicOnly is too high so I would try removing it

```python
lm2_ex2 = smf.ols(formula = 'TotalIncome~TotalPop+PrivateOnly+WithBothPublicAndPrivate',data=ex2).fit()
```

```python
modelEval(lm2_ex2)
```

### Now the impact of the Total Population, Private Only and both Private and Public are statistically significant (p-value < 5%) and it seems that the private only pop has the most impact. In addition, the validation R2 is improved so the fit is good

### (c) Try to fit y with polynomial of x with order 1, 2, 3, 4, respectively
 For each model, store your fitted parameters as well as your $R^2$

```python
n = len(ex3.columns)-1
```

```python
def AdjR2(flag):
    '''
    Evalute the model with adjusted R-squared,
    with flag indicating the feature selection
    1--feature included
    0--feature excluded
    '''
    formula = 'y~'
    for i in range(len(flag)):
        if flag[i] == 1:
            formula += '+x{}'.format(i+1)
    train, valid = train_test_split(ex3, test_size = 0)
    lm = smf.ols(formula = formula , data = train).fit()
    lmy = lm.predict(valid)
    y_err = lmy-valid.y
    y_norm = valid.y-mean(valid.y)
    # Adjusted R^2
    R2 = 1 - y_err.dot(y_err) / y_norm.dot(y_norm) * (n-1) / (n-sum(flag)-1)
    return R2, lm
```

```python
adjR2 = []
regR2 = []
for i in range(1, n+1):
    a,b = AdjR2([1]*i + [0]*(n-i))
    adjR2.append(a)
    regR2.append(b.rsquared)
```

### (d) Report the  $R^2$ , plot it and observe how it changes

```python
regR2
```

```python
pl.figure(figsize=[16,5])
pl.plot(range(1,n+1),regR2,'b-*',markersize=10,label='$R^2$')
pl.legend()
pl.show()
```

### (e) Create a figure with four (4) subfigures, one for each polynomial degree M = 1, 2, 3, 4. In each subfigure:
    * (e.1) Draw the original scatter plot of x1 and y
    * (e.2) Plot the curve of true function $f(x) = 9x^2 - x + 1$
    * (e.3) Plot your fitted curve of degree M, observe how our model fits the data as M increases

```python
ex3
```

```python
fig = pl.figure(figsize=[16,16])
g = lambda x: (9*(x**2) - x +1)
for M in [1,2,3,4]:
    ax = fig.add_subplot(2,2,M)
    ax.scatter(ex3.x1,ex3.y,label='original x1 & y',c='r')
    polyfit = np.polyfit(ex3.x1, ex3.y, M)
    f = np.poly1d(polyfit)
    x_new = np.linspace(ex3.x1.min(), ex3.x1.max(), 10000)
    y_new = f(x_new)
    lmy = f(ex3.x1)
    y_err = lmy - ex3.y
    y_norm = ex3.y - np.mean(ex3.y)
    R2 = 1-y_err.dot(y_err)/y_norm.dot(y_norm)
    ax.plot(x_new,g(x_new),label='true function',c='b')
    ax.plot(x_new,y_new,label='polynomial fit',c='g')    
    ax.legend()
    formula = 'y = '+ str(round(f.coeffs[-1],4)) 
    for i in range(len(f),0,-1):
        formula = formula + ' + '+ str(round(f.coeffs[len(f)-i],4)) + '$x^'+str(i)+'$'
    ax.set_title('M = {}\n $R^2$ = {} \n {}'.format(M,R2,formula))
fig.suptitle('Fitting for y and x1', y = 0.92,fontsize=16)
```

### It seems that based on plot, polynomial model with degree of freedom 2 fits the true curve best

```python
fig = pl.figure(figsize=[20,20])
g = lambda x: (9*(x**2) - x +1)
for M in [1,2,3,4]:
    ax = fig.add_subplot(2,2,M)
    ax.scatter(ex3.x1,ex3.y,label='original x1 & y',c='r')
    polyfit = np.polyfit(ex3.x1, ex3.y, M)
    f = np.poly1d(polyfit)
    np.random.seed(999)
    prediction_error = []
    yi = []
    xi = []
    fxi = []
    for i in range(10):
        x0 = np.random.uniform(np.min(ex3.x1),np.max(ex3.x1))
        ax.plot(x0,g(x0),'yd',markersize=12,label='x picked on true function')
        ax.plot(x0,f(x0),'kd',markersize=12,label='x picked on fitted function')
        ax.plot([x0,x0],[g(x0),f(x0)],'--',label='prediction error')
        prediction_error.append(-g(x0)+f(x0))
        xi.append(x0)
        yi.append(g(x0))
        fxi.append(f(x0))
    x_new = np.linspace(ex3.x1.min(), ex3.x1.max(), 10000)
    y_new = f(x_new)
    lmy = f(ex3.x1)
    y_err = lmy - ex3.y
    y_norm = ex3.y - np.mean(ex3.y)
    R2 = 1-y_err.dot(y_err)/y_norm.dot(y_norm)
    prediction_error = np.mean(prediction_error)
    E = np.sum((np.array(fxi) - np.array(yi))**2)/10
    ax.plot(x_new,g(x_new),label='true function',c='b')
    ax.plot(x_new,y_new,label='polynomial fit',c='g')    
    # Make label only one
    handles, labels = pl.gca().get_legend_handles_labels()
    by_label = OrderedDict(zip(labels, handles))
    pl.legend(by_label.values(), by_label.keys())
    formula = 'y = '+ str(round(f.coeffs[-1],4)) 
    
    for i in range(len(f),0,-1):
        formula = formula + ' + '+ str(round(f.coeffs[len(f)-i],4)) + '$x^'+str(i)+'$'
    ax.set_title('M = {}\n $R^2$ = {} \n {}\n Mean Prediction Error = {}\n E = {}'.format(M,R2,formula,prediction_error,E))
fig.suptitle('Fitting for y and x1', y = 0.92,fontsize=16)
```
